```python
import pandas as pd
```

```python
movies=pd.read_csv("/content/drive/MyDrive/IR project/mcu-movies.csv")
movies.dropna(inplace=True)
movies
```

```python
movies['movie']=movies['movie'].str.strip()
#movies['movie']=movies['movie'].str.replace(" ", "")
movie_list=movies['movie'].to_list()
len(movie_list)
```

```python
characters=pd.read_csv('/content/drive/MyDrive/IR project/marvel-wikia-data.csv')
characters
```

```python
characters["name"] = characters["name"].str.split('(').str[0]
characters["name"] = characters["name"].str.strip()
character_list=characters["name"].to_list()
character_list

```

```python
for i in character_list :
    movie_list.append(i)
```

```python
len(movie_list)
```

```python
movie_list=list(set(movie_list))
```

```python
len(movie_list)
```

```python
# Import packages
import wikipedia
title=[]
texts=[]
urls=[]
summary=[]


for i in movie_list:
  try:
      # Specify the title of the Wikipedia page
      wiki = wikipedia.page(i,auto_suggest=True, redirect=True)

      # Extract the plain text content of the page, excluding images, tables, and other data.
      text = wiki.content

      # Replace '==' with '' (an empty string)
      text = text.replace('==', '')

      # Replace '\n' (a new line) with '' & end the string at $1000.
      text = text.replace('\n', '')[:-12]
      texts.append(text)
      title.append(i)
      summary.append(wikipedia.summary(i))
      urls.append(wiki.url)
  except:
      pass  

df=pd.DataFrame({'Title':title,'Summary':summary,'Text':texts,'URL':urls })
df.to_csv('/content/drive/MyDrive/IR project/marvel_wiki.csv')
```
